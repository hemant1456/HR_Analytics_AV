```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
```

```python
df= pd.read_csv('train_LZdllcl.csv',index_col=0)
```

```python
fig= plt.figure(figsize=(18,12))
gs= gridspec.GridSpec(3,3,figure=fig)


# Define the subplot locations
ax1 = plt.subplot(gs[0, :2])  # spans 2 columns
ax2 = plt.subplot(gs[0:2, 2])
ax3 = plt.subplot(gs[1, 0])
ax4 = plt.subplot(gs[1, 1])
#ax5 = plt.subplot(gs[1, 2])
ax6 = plt.subplot(gs[2, 0])
ax7 = plt.subplot(gs[2, 1])
ax8 = plt.subplot(gs[2, 2])


sns.countplot(x=df['department'],hue=df['is_promoted'],ax=ax1)
labels = [label.get_text().replace(' ', '\n') for label in ax1.get_xticklabels()]
ax1.set_xticklabels(labels)
plt.tick_params(axis='x', labelsize=12)
sns.countplot(x=df['gender'],hue=df['is_promoted'],ax=ax2)
sns.countplot(x=df['recruitment_channel'],hue=df['is_promoted'],ax=ax3)
sns.countplot(x=df['no_of_trainings'],hue=df['is_promoted'],ax=ax4)
ax4.legend(loc='upper right')
#sns.countplot(x=df['KPIs_met >80%'],hue=df['is_promoted'],ax=ax5)
sns.countplot(x=df['education'],hue=df['is_promoted'],ax=ax6)
sns.countplot(x=df['awards_won?'],hue=df['is_promoted'],ax=ax7)
sns.countplot(x=df['previous_year_rating'],hue=df['is_promoted'],ax=ax8)
```
